# Filling Missing Values

```python
for i in train_df.select_dtypes(include="float64"):
    train_df[i].fillna(train_df[i].median(),inplace=True)
    
for i in test_df.select_dtypes(include="float64"):
    test_df[i].fillna(test_df[i].median(),inplace=True)
```

```python
# DATA CLEANING
train_df["Ticket"].unique()
```

```python
test_df["Ticket"].unique()
```

```python
l=[]
for i in test_df["Ticket"]:
    ticket=i.split(" ")
    if len(ticket)==1:
        l.append(ticket[0])
    elif len(ticket)==2:
        l.append(ticket[1])
    elif len(ticket)==3:
        l.append(ticket[2])
test_df["Ticket"]=l
test_df["Ticket"]=test_df["Ticket"].astype("int32")
```

```python
test_df["Ticket"].unique()
```

```python
# DATA PREPROCESSING

train_df["Embarked"].replace(["S","C","Q"],["Southampton","Cherbourg","Queenstown"],inplace=True)
conditions = [
    ((train_df["Age"]>0) & (train_df["Age"]<13.0)),
    (train_df["Age"] >=13.0) & (train_df["Age"] <18.0),
    (train_df["Age"]>=18.0) & (train_df["Age"]<32.0),
    (train_df["Age"]>=32.0)& (train_df["Age"]<50.0),
    (train_df["Age"]>=50.0)]
values = ["Child","Teenage","Young","Mid-Age","Senior"]
train_df['Age Group'] = np.select(conditions, values)
train_df["Family"]=train_df["SibSp"]+train_df["Parch"]

test_df["Embarked"].replace(["S","C","Q"],["Southampton","Cherbourg","Queenstown"],inplace=True)
conditions = [
    ((test_df["Age"]>0) & (test_df["Age"]<13.0)),
    (test_df["Age"] >=13.0) & (test_df["Age"] <18.0),
    (test_df["Age"]>=18.0) & (test_df["Age"]<32.0),
    (test_df["Age"]>=32.0)& (test_df["Age"]<50.0),
    (test_df["Age"]>=50.0)]
values = ["Child","Teenage","Young","Mid-Age","Senior"]
test_df['Age Group'] = np.select(conditions, values)
test_df["Family"]=test_df["SibSp"]+test_df["Parch"]
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(train_df["Sex"])
train_df["Sex"]=le.transform(train_df["Sex"])
le.fit(train_df["Age Group"])
train_df["Age Group"]=le.transform(train_df["Age Group"])

le.fit(test_df["Sex"])
test_df["Sex"]=le.transform(test_df["Sex"])
le.fit(test_df["Age Group"])
test_df["Age Group"]=le.transform(test_df["Age Group"])
```

```python
x_train=train_df[["Pclass","Sex","Age","SibSp","Parch","Fare","Age Group","Family"]]
y_train=train_df["Survived"]

x_test=test_df[["Pclass","Sex","Age","SibSp","Parch","Fare","Age Group","Family"]]
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
ro=RandomOverSampler()
train_inp,train_out=ro.fit_resample(x_train,y_train)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier(criterion="gini",splitter="best")
dc.fit(train_inp,train_out)
train=dc.score(train_inp,train_out)*100
print("Traing Accuracy: ",train)
```

```python
y_test=dc.predict(x_test)
pred_df=pd.DataFrame({"PassengerID":test_id,"Survived":y_test})
pred_df
```
